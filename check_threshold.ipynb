```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
# Load audio file into NumPy array
sample_rate, data = wavfile.read('/content/drone+helicopter_01.wav')

# Compute FFT of the audio signal
fft_data = np.fft.rfft(data)

# Compute power spectrum and frequency bins
power_spectrum = np.abs(fft_data) ** 2
frequency_bins = np.linspace(0, sample_rate / 2, len(power_spectrum))

# Calculate mean and standard deviation of power spectrum
mean_power = np.mean(power_spectrum)
std_power = np.std(power_spectrum)

# Define threshold as multiple of standard deviation above the mean
threshold = mean_power + 3 * std_power

# Check if power spectrum values exceed the threshold
if np.max(power_spectrum) > threshold:
    print("Drone audio present")
else:
    print("No drone audio detected")

# Visualize the power spectrum
plt.plot(frequency_bins, power_spectrum)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power')
plt.axhline(threshold, color='r')
plt.show()
print(threshold)
```
